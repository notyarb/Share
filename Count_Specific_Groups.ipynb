```python
import pandas as pd
from rdkit import Chem

df = pd.read_csv("Composition of Human Urine.csv") 

dfList = list(df['InChi'])
dfListstring = []
for s in dfList: 
    inchi = str(s)
    dfListstring.append(inchi)
    
inchis = ['InChI=' + s for s in dfListstring]
inchis.remove('InChI=nan')
inchis.remove('InChI=nan')
inchis.remove('InChI=nan')
#print(inchis)

for i in inchis:
    mol = Chem.MolFromInchi(i)
    functional_group1 = Chem.MolFromSmarts('C=O')
    matches1 = mol.GetSubstructMatches(functional_group1)
    print ('Index Number:' + str(inchis.index(i)))
    print ('Number of C=O')
    print (len(matches1))
    print (' ')
```
